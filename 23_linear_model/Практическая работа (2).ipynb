**Задача 4. Стратегия голосования**

Реализуйте стратегию голосования для предсказания целевой переменной.
Голосование в задаче классификации — это когда несколько моделей выдают свои предикты, и финальным выбирается тот предикт, который предсказали большинство моделей.

Для реализации этой стратегии проделайте следующее:

- сохраните предсказания каждой из моделей (случайный лес, многослойный персептрон, логистическая регрессия) для тестовой выборки в датафрейм `pred_df`;
- в четвертую колонку `target` положите тот класс, который предсказало большинство классификаторов. Например, если в строке были значения `high, medium, medium`, в `target` нужно положить `medium`.

     Если в строке три разных класса — `high, medium, low`, — придумайте свою стратегию по выбору значения. Самая простая стратегия: выбрать рандомно одно значение из трёх.

- посчитайте точность предсказания с помощью голосования. Выведите значения метрик accuracy и confusion_matrix.


Добейтесь значения точности > 78% 

```python
pred_df1 = pd.DataFrame({
    'rf' : predicted_test_rf, 
    'mlp' : predicted_test_mlp, 
    'lr' : logreg_pred,
})
pred_df2 = pred_df1.copy()
pred_df2.head(20)
```

```python
import statistics
#количество раз, когда данный метод проигрывал в голосовании
stats = {'rf' : 0, 'mlp' : 0, 'lr' : 0} 
def target_choice1(rf, mlp, lr):
    names = {'rf' : rf, 'mlp': mlp, 'lr' : lr}
    #если все значения одинаковые
    if (rf == mlp) & (rf == lr) & (mlp == lr):
        return rf
    #если хотя бы одно значение отличается
    elif (rf == mlp) | (rf == lr) | (mlp == lr):
        #ищем значение, отличное от двух других
        not_mode = lambda x,y,z: [i for i in [x, y, z] if i != statistics.mode([x, y, z])]
        not_target_pred = not_mode(rf, mlp, lr)
        #ищем метод, который выдал отличное значение
        key = next((key for key, i in names.items() if i == not_target_pred[0]), None)
        #записываем в словарь количество ошибок метода
        stats[key] += 1
        #записываем в target значение, победившее в голосовании
        return statistics.mode([rf, mlp, lr])
    else:
        #если все значения отличны, выбираем то, которое выдал метод с 
        #наименьшим количеством ошибок 
        return names[min(stats, key = stats.get)]
```

```python
import statistics
import random
def target_choice2(rf, mlp, lr):
    #если все значения одинаковые
    if (rf == mlp) & (rf == lr) & (mlp == lr):
        return rf
    #если хотя бы одно значение отличается
    elif (rf == mlp) | (rf == lr) | (mlp == lr):
        return statistics.mode([rf, mlp, lr])
    else:
        #если все значения отличны, выбираем рандомно
        return random.choice([rf, mlp, lr])
```

```python
target1 = []
target2 = []

for i in range(len(pred_df1)):
    target1.append(pd.Series(target_choice1(pred_df1.rf[i], pred_df1.mlp[i], pred_df1.lr[i]))[0])
for i in range(len(pred_df2)):
    target2.append(pd.Series(target_choice2(pred_df2.rf[i], pred_df2.mlp[i], pred_df2.lr[i]))[0])
```

```python
pred_df1['target'] = target1
pred_df2['target'] = target2
#pred_df[(pred_df.rf != pred_df.mlp) & (pred_df.rf != pred_df.lr) & (pred_df.mlp != pred_df.lr)]
```
