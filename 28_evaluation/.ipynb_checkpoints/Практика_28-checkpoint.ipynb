#### Аномалии в стоимости


```python
#Ящик с усами для распределения стоимости автомобиля
fig, ax = plt.subplots(figsize=(5,15))
ax.boxplot(df.price)
ax.set_xlabel('Стоимость автомобиля')
ax.set_ylabel('Частота')
ax.set_title('Распределение стоимости автомобиля')

ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
```

```python
#Описательные статистики для стоимости автомобиля
df.price.describe().apply(lambda x: f"{x:0.3f}")
```

```python
#Интерквартериальный размах
def calculate_outliers(data):
    q25 = data.quantile(0.25)
    q75 = data.quantile(0.75)
    iqr = q75 - q25
    bounds = (q25 - 1.5 * iqr, q75 + 1.5 * iqr)
    return bounds
```

```python
#Границы размаха
bounds = calculate_outliers(df.price)
bounds
```

```python
#Подсчет количества аномалий
is_outlier = (df.price < bounds[0]) | (df.price > bounds[1])
is_outlier.sum()
```

```python
#Присвоение аномалиям верхнего граничного значения и сохранение результата в новый датафрейм
df_clean = df.copy()
df_clean.price[df_clean.price > bounds[1]] = bounds[1]
df_clean.head(5)
```

```python
#Ящик с усами для распределения стоимости автомобиля
fig, ax = plt.subplots(figsize=(5,15))
ax.boxplot(df_clean.price)
ax.set_xlabel('Стоимость автомобиля')
ax.set_ylabel('Частота')
ax.set_title('Распределение стоимости автомобиля')

ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
```

#### Стандартизация

```python
#Стандартизация 
vars = ['odometer_km', #cоздание списка переменных для стандартизации
        'odometer/price', 
        'listing_age', 
        'len_desc', 
        'month', 
        'dayofweek', 
        'diff_years', 
        'model_in_desc', 
        'model_len', 
        'model_words_count', 
        'lat', 
        'long', 
        'year']
std_scaler = StandardScaler() #создание объекта типа StandardScaler
std_scaler.fit(df_clean[vars]) #идентификация переменных для стандартизации
scaled= std_scaler.transform(df_clean[vars]) #трансформация переменных
vars_std_scaled = [] #создание списка с иенами новых переменных
for i in vars:
    vars_std_scaled.append(i + '_std')
df_clean[vars_std_scaled] = scaled #запись новых переменных в датафрейм
df_clean.head()
```

#### Удаление неинформативных колонок

```python
columns_for_drop = ['year', 'url', 'region', 'region_url', 'manufacturer',
                    'model', 'fuel', 'odometer', 'title_status', 'transmission',
                    'image_url', 'description', 'state', 'lat', 'long', 'posting_date',
                    'odometer_km', 'odometer/price',
                    'len_desc', 'model_in_desc', 'model_len', 'model_words_count',
                    'short_model', 'date',
                    'month', 'dayofweek', 'diff_years',
                    'listing_age']
df_prepared = df_clean.drop(columns_for_drop, axis = 1)
df_prepared.isna().sum()
```

## Modeling

```python
#Инициализация фичей и целевой переменной
x = df_prepared.drop(['price_category', 'id'], axis=1)
y = df_prepared['price_category']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#Объявление моделей разных типов
logreg = LogisticRegression(random_state=42) #логистическая регрессия
randfor = RandomForestClassifier(random_state=42) #случайный лес
mlp = MLPClassifier(random_state=42) #многослойный персептрон 
```

```python
#logreg.fit(x_train, y_train)
```

```python
#Кросс-валидация для логистической регрессии с разделением исходного сэмпла на 5 частей
cv_score_lr = cross_val_score(logreg, x_train, y_train, cv=5)
cv_score_lr
```

```python
#Среднее и стандартное отклонение 
cv_score_lr.mean(), cv_score_lr.std()
```

```python
#Кросс-валидация для случайного леса с разделением исходного сэмпла на 5 частей
cv_score_rf = cross_val_score(randfor, x_train, y_train, cv=5)
cv_score_rf
```

```python
#Среднее и стандартное отклонение 
cv_score_rf.mean(), cv_score_rf.std()
```

```python
#Кросс-валидация для многослойного персептрона с разделением исходного сэмпла на 5 частей
cv_score_mlp = cross_val_score(mlp, x_train, y_train, cv=5)
cv_score_mlp
```

```python
#Среднее и стандартное отклонение 
cv_score_mlp.mean(), cv_score_mlp.std()
```

```python
#Рандомная сетка для поиска лучших параметров случайного леса
param_grid = {
    'bootstrap':[True, False],
    'max_depth' : list(range(1,100,5)),
    'max_features': ['sqrt', 'log2'],
    'min_samples_leaf': list(range(1,10)),
    'min_samples_split':list(range(2,10)),
    'random_state': list(range(1,100)),
    'n_estimators':list(range(50, 501, 100))
}
randomized_search_rf = RandomizedSearchCV(
   estimator=randfor,
   param_distributions=param_grid,
   n_iter=20,
   scoring='accuracy',
   verbose=1,
   n_jobs=-1
)
randomized_search_rf.fit(x_train, y_train)
best_params_rf= randomized_search_rf.best_params_
print(best_params_rf)
```

```python
#Запуск случайного леса с наилучшими параметрами
randfor_tuned = RandomForestClassifier(
    random_state= best_params_rf['random_state'], 
    n_estimators= best_params_rf['n_estimators'],
    min_samples_split= best_params_rf['min_samples_split'],
    min_samples_leaf= best_params_rf['min_samples_leaf'],
    max_features= best_params_rf['max_features'], 
    max_depth= best_params_rf['max_depth'], 
    bootstrap= best_params_rf['bootstrap']
)
cv_score_rf = cross_val_score(randfor, x, y, cv=5)
cv_score_rf.mean(), cv_score_rf.std()
```

```python
#Предсказание случайного леса на тестовой части сэмпла
randfor_tuned.fit(x_train, y_train)

predicted_train_rf_tuned = randfor_tuned.predict(x_train)
predicted_test_rf_tuned = randfor_tuned.predict(x_test)

print(accuracy_score(y_train, predicted_train_rf_tuned))
print(accuracy_score(y_test, predicted_test_rf_tuned))
```

```python
#Коэффициенты важности для случайного леса
f_imp_list_rf = list(zip(x_train.columns, randfor_tuned.feature_importances_))
f_imp_list_rf.sort(key=lambda x: x[1], reverse=True)
f_imp_list_rf
```

```python
#Рандомная сетка для поиска лучших параметров многослойного персептрона
param_grid = {
    'activation' : ['identity', 'logistic', 'tanh', 'relu'],
    'max_iter': list(range(200,600,50)),
    'solver': ['lbfgs', 'sgd', 'adam']
}
randomized_search_mlp = RandomizedSearchCV(
   estimator=mlp,
   param_distributions=param_grid,
   n_iter=10,
   scoring='accuracy',
   verbose=1,
   n_jobs=-1
)
randomized_search_mlp.fit(x_train, y_train)
best_params_mlp = randomized_search_mlp.best_params_
print(best_params_mlp)
```

```python
#Запуск многослойного персептрона с наилучшими параметрами
mlp_tuned = MLPClassifier(
    activation= best_params_mlp['activation'],
    max_iter= best_params_mlp['max_iter'],
    solver= best_params_mlp['solver'],
)
cv_score_mlp = cross_val_score(mlp, x, y, cv=5)
cv_score_mlp.mean(), cv_score_mlp.std() #среднее значение и отклонение
```

```python
#Предсказание многослойного персептрона на тестовой части сэмпла
mlp_tuned.fit(x_train, y_train)

predicted_train_mlp_tuned = mlp_tuned.predict(x_train)
predicted_test_mlp_tuned = mlp_tuned.predict(x_test)

print(accuracy_score(y_train, predicted_train_mlp_tuned))
print(accuracy_score(y_test, predicted_test_mlp_tuned))
```

```python
#Рандомная сетка для поиска лучших параметров логистической регрессии
param_grid = {
    'C': [1,1.5,2,2.5,3],
    'solver' : ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga'],
    'max_iter': list(range(100,700,50)),
}
randomized_search_lr = RandomizedSearchCV(
   estimator=logreg,
   param_distributions=param_grid,
   n_iter=20,
   scoring='accuracy',
   verbose=1,
   n_jobs=-1
)
randomized_search_lr.fit(x_train, y_train)
best_params_lr = randomized_search_lr.best_params_
print(best_params_lr)
```

```python
#Запуск логистической регрессии с наилучшими параметрами
logreg_tuned = LogisticRegression(
    C= best_params_lr['C'], 
    solver= best_params_lr['solver'],
    max_iter= best_params_lr['max_iter'],
)
cv_score_lr = cross_val_score(logreg, x, y, cv=5)
cv_score_lr.mean(), cv_score_lr.std()
```

```python
#Предсказание логистической регрессии на тестовой части сэмпла
logreg_tuned.fit(x_train, y_train)

predicted_train_lr_tuned = logreg_tuned.predict(x_train)
predicted_test_lr_tuned = logreg_tuned.predict(x_test)

print(accuracy_score(y_train, predicted_train_lr_tuned))
print(accuracy_score(y_test, predicted_test_lr_tuned))
```
